```python
import pandas as pd
forbes = pd.read_csv("FullSetUncombined/nvda_sentence_sentiment_dataset_forbes_cleaned.csv")
motley = pd.read_csv("FullSetUncombined/nvda_sentence_sentiment_dataset_motley_cleaned.csv")
```

```python
print(forbes.head())
print(forbes.shape[0])
print(motley.head())
print(motley.shape[0])
```

```python

combined_df = pd.concat([forbes, motley], axis=0, ignore_index=True)
print(combined_df.shape[0])
```

```python
print(combined_df.isna().sum())
```

```python
# Combine each pair of columns: use values from 'score_*' when 'positive_prob', etc. are missing
combined_df['positive_prob'] = combined_df['positive_prob'].combine_first(combined_df['score_positive'])
combined_df['neutral_prob'] = combined_df['neutral_prob'].combine_first(combined_df['score_neutral'])
combined_df['negative_prob'] = combined_df['negative_prob'].combine_first(combined_df['score_negative'])

# Optionally drop the misnamed columns once combined
combined_df.drop(columns=['score_positive', 'score_neutral', 'score_negative'], inplace=True)

# Check for any remaining missing values
print(combined_df.isna().sum())
```

```python
# Random sample 50 rows from the combined DataFrame
sampled_df = combined_df.sample(n=50, random_state=42)
sampled_df.head(50)
```

```python
combined_df.to_csv("full_set.csv", index=False)
print("Files saved:\n- FullSet/full_set.csv (full dataset)")
```
